# Usage guide for visualization functions

This notebook provides a simple usage guide for the visualization functions implemented in `metalhistory/visualization_api`.

Start with the imports...

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

from PIL import Image

import metalhistory.visualization_api as vis
```

## Artist barplot

The `artist_barplot` plots statistics of artists with at least a certain number of albums produced. The statistics include min/mean/max and can be computed based on *unique listeners*, *total playcount* or *MA score*.

```python
file_name = 'test_barplot.png'
vis.artist_barplot(min_albums=5, n_artists=30, metric='MA_score', file_name=file_name)
img = Image.open(file_name)
ax = plt.imshow(np.asarray(img))
ax.figure.set_size_inches(12, 12)
```

## Artist word cloud

The `artist_cloud` function visualizes the most popular artists by drawing a wordcloud with their names.

```python
file_name = 'test_wordcloud.png'
vis.artist_cloud(min_albums=5, words_limit=20, metric='MA_score', file_name=file_name)
img = Image.open(file_name)
ax = plt.imshow(np.asarray(img))
ax.figure.set_size_inches(12, 12)
```

## Album covers

To visualize album covers based on playcount use `album_covers`. This function reads the album covers of the most popular albums and creates an image such that the area of each album scales with the playcount. (To save the image to disc set `image_name` to the name of the desired output file.)

```python
img = vis.album_covers(num_albums=100, width=1280, height=720, image_name=None)
ax = plt.imshow(np.asarray(img))
ax.figure.set_size_inches(12, 12)
```
